```python
def sig_stars(pval):
    if pval < 1e-2:
        num_stars = 3
    elif pval < 5e-2:
        num_stars = 2
    elif pval < 1e-1:
        num_stars = 1
    else:
        num_stars = 0
        
    return(num_stars)
```

```python
def make_est_str(r):
    try:
        est_str = str(round(r['estimate'], 2))
        ci_range = round(abs(r['estimate'] - r['conf.low']), 2)
        
        if not np.isnan(ci_range):
            est_str = f'{est_str}(±{ci_range})'
            est_str += sig_stars(r['p.value'])*'*'
        
    except Exception:
        est_str = np.nan

    return(est_str)
    
```
